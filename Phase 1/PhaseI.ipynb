## Downloaded Dataset Requirement



```python
import pandas as pd
def data_parser():
    df = pd.read_csv('fueleconomydatacsv.csv', low_memory = False)
    x = df.iloc[2534:,:]
    for col in x.columns:
        if not col in ["Model Year", "Division", "Carline", "Comb FE (Guide) - Conventional Fuel"]:
            del x[col]
    x.to_csv('carFE.csv', index = False, header = True)
    return x
############ Function Call ############
data_parser()
```

## Web Collection Requirement \#1


```python
import re
import requests
from bs4 import BeautifulSoup
from pprint import pprint
import pandas as pd
import numpy as np
names  = []
def web_parser1():
    url = 'https://www.goodcarbadcar.net/2020-us-vehicle-sales-figures-by-model/'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    tables = soup.find_all("table")
    table = tables[0]
    x = table.find_all("tr")
    df = pd.DataFrame()
    
    recent =[]
    past = []
    data1 =[]
    for i in x:
        if len(i.find_all("td")) != 0:
            name = i.find_all("td")[0].get_text()
            recentsales = i.find_all("td")[1].get_text()
            pastsales = i.find_all("td")[2].get_text()
            if name != "modelName":
                names.append(name)
                recent.append(recentsales)
                past.append(pastsales)

    for i in range(len(names)):
        df.loc[names[i], "2019 Q3 Sales"] = recent[i]
        df.loc[names[i], "2020 Q3 Sales"] = past[i]
    
    
    df.to_csv('carsales.csv', index = True, header = True)
    return df
        
############ Function Call ############
web_parser1()
```

## Web Collection Requirement \#2

```python
import re
import requests
from bs4 import BeautifulSoup
from pprint import pprint
import pandas as pd
import numpy as np
import json
def web_parser2():
    ratingdf  = pd.DataFrame()
    for i in range(len(names)):
        if len(names[i].split()) < 2:
            continue
        make  = names[i].split()[0]
        model = names[i].split()[1]
        recenturl = "https://www.jdpower.com/cars/2020/"+ make + "/" + model
        response = requests.get(recenturl)
        soup = BeautifulSoup(response.text, "html.parser")
        h = soup.find_all("script")[2]
        try:
            recent_rating = json.loads(h.decode_contents())['review']['reviewRating']['ratingValue']
            
        except: 
            recent_rating = None
            
        ratingdf.loc[names[i],"2020 rating"] = recent_rating
        
        
        
        pasturl = "https://www.jdpower.com/cars/2019/"+ make + "/" + model
        response1 = requests.get(pasturl)
        soup1 = BeautifulSoup(response1.text, "html.parser")
        h1 = soup1.find_all("script")[2]
        try:
            past_rating = json.loads(h1.decode_contents())['review']['reviewRating']['ratingValue']    
        except:
            past_rating = None
        ratingdf.loc[names[i],"2019 rating"] = past_rating
        
        
    ratingdf.to_csv('carrating.csv', index = True, header = True)
    
    return ratingdf
    
            


############ Function Call ############
web_parser2()
```
